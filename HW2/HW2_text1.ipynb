```python
import math

# 物理与化学常数
N_A = 6.022e23       # 阿伏伽德罗常数, mol^-1
M_Bi = 208.98        # 铋的相对原子质量, g/mol

# 样品参数
m_soil = 20e-3       # 月壤样本质量, g
c_Bi = 3e-9          # 铋的质量分数 (3 ppb)

# 计算样品中 Bi 的质量并求靶核数量 N_0
m_Bi = m_soil * c_Bi
N_0 = (m_Bi / M_Bi) * N_A
print(f"靶核总数 N_0: {N_0:.4e}")
```

```python
# 核反应堆与核反应参数
phi = 5e14           # 中子通量, n/(cm^2*s)
sigma = 24e-27       # 基态生成截面, 24 mb = 24e-27 cm^2
t_half = 5.01        # 210Bi 基态半衰期, days
t_i = 5.0            # 辐照时间, days

# 计算衰变常数
lambda_decay = math.log(2) / t_half
print(f"衰变常数 lambda: {lambda_decay:.4f} days^-1")
```

```python
# 计算活化饱和因子
saturation_factor = 1 - math.exp(-lambda_decay * t_i)

# 计算辐照结束时的活度 (单位: Bq)
A_Bq = N_0 * phi * sigma * saturation_factor

# 转换活度单位为 dpm (每分钟衰变数)
A_dpm = A_Bq * 60

# 计算预期计数率 cpm (探测效率为 25%)
efficiency = 0.25
count_rate_cpm = A_dpm * efficiency

print(f"饱和因子: {saturation_factor:.4f}")
print(f"初始活度 A0: {A_Bq:.4f} Bq")
print(f"初始活度 A0: {A_dpm:.4f} dpm")
print(f"预期探测计数率: {count_rate_cpm:.4f} cpm")
```
